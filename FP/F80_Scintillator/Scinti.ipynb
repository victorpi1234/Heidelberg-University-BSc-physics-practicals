Find endpoints by finding last point with more than 10 counts

```python
def find_last_above_10(filepath, max_channels=1500):
    with open(filepath, 'r') as f:
        data = f.read()
    # Convert comma decimals and split into numbers
    numbers = [float(x.replace(',', '.')) for x in data.split() if x]
    # Only consider first max_channels points
    numbers = numbers[:max_channels]
    # Search backwards for first number > 10
    for i in range(len(numbers)-1, -1, -1):
        if numbers[i] > 9:
            return i
    return None  # Return None if no value >10 found

# File data
file_data = {
    "0.5": "4.3.3/Sr90_4min_CG_32_Al_0.5mm_raw.dat",
    "1.0": "4.3.3/Sr90_4min_CG_32_Al_1mm_raw.dat",
    "1.5": "4.3.3/Sr90_4min_CG_32_Al_1.5mm_raw.dat",
    "2.0": "4.3.3/Sr90_4min_CG_32_Al_2mm_raw.dat",
    "2.5": "4.3.3/Sr90_4min_CG_32_Al_2.5mm_raw.dat",
    "3.0": "4.3.3/Sr90_4min_CG_32_Al_3mm_raw.dat",
    "3.5": "4.3.3/Sr90_4min_CG_32_Al_3.5mm_raw.dat",
    "4.5": "4.3.3/Sr90_4min_CG_32_Al_4.5mm_raw.dat"
}

# Process all files
thicknesses = []
channels = []

for thickness_str, filepath in file_data.items():
    channel = find_last_above_10(filepath, max_channels=600)
    if channel is not None:
        thicknesses.append(float(thickness_str))
        channels.append(channel)
        print(f"{thickness_str}mm: Last channel >10 (first 700) = {channel}")

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(thicknesses, channels, 'bo-', markersize=8)
plt.xlabel('Aluminum Thickness (mm)', fontsize=12)
plt.ylabel('Channel Number (first 700)', fontsize=12)
plt.title('Last Channel with Counts >10 (First 700 Channels)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(thicknesses)
plt.tight_layout()

# Save and show the plot
plt.savefig('channel_vs_thickness_first700.png', dpi=300)
plt.show()
```

Find end points by kurie plots

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def f80_readfile(filepath):
    with open(filepath, 'r') as f:
        data = f.read()
    return np.array([float(x.replace(',', '.')) for x in data.split() if x])

def analyze_file(filepath, fit_range, force_zero=True):
    ydata = f80_readfile(filepath)
    xdata = np.arange(len(ydata))
    
    # Select fit region
    mask = (xdata >= fit_range[0]) & (xdata <= fit_range[1])
    x_fit = xdata[mask]
    y_fit = ydata[mask]
    
    if force_zero:
        # Modified linear fit constrained to approach zero
        def fit_func(x, a):
            return a * (x - fit_range[1])  # Force zero at fit_range[1]
        
        try:
            a, = curve_fit(fit_func, x_fit, y_fit)
            endpoint = fit_range[1]  # Where we forced y=0
            fit_line = a * (xdata - endpoint)
        except:
            # Fallback to normal linear fit if constrained fails
            m, b = np.polyfit(x_fit, y_fit, 1)
            endpoint = -b/m if m != 0 else 0
            fit_line = m * xdata + b
    else:
        # Standard linear fit
        m, b = np.polyfit(x_fit, y_fit, 1)
        endpoint = -b/m if m != 0 else 0
        fit_line = m * xdata + b
    
    return endpoint, xdata, ydata, fit_line

# Files with individual fit ranges
file_data = [
    {"path": "4.3.3/Sr90_4min_CG_32_Al_0.5mm_raw.dat", "fit_range": (300, 600)},
    {"path": "4.3.3/Sr90_4min_CG_32_Al_1mm_raw.dat", "fit_range": (250, 400)},
    {"path": "4.3.3/Sr90_4min_CG_32_Al_1.5mm_raw.dat", "fit_range": (200, 380)},
    {"path": "4.3.3/Sr90_4min_CG_32_Al_2mm_raw.dat", "fit_range": (190, 330)},
    {"path": "4.3.3/Sr90_4min_CG_32_Al_2.5mm_raw.dat", "fit_range": (180, 320)},
    {"path": "4.3.3/Sr90_4min_CG_32_Al_3mm_raw.dat", "fit_range": (170, 330)},
    {"path": "4.3.3/Sr90_4min_CG_32_Al_3.5mm_raw.dat", "fit_range": (190, 320)},
    {"path": "4.3.3/Sr90_4min_CG_32_Al_4.5mm_raw.dat", "fit_range": (180, 320)}
]

thicknesses = [0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.5]
endpoints = []

plt.figure(figsize=(12, 8))
for i, file_info in enumerate(file_data):
    endpoint, x, y, fit = analyze_file(file_info["path"], file_info["fit_range"])
    endpoints.append(endpoint)
    
    plt.subplot(2, 4, i+1)
    plt.plot(x, y, 'b-', label='Data')
    plt.plot(x, fit, 'r-', label=f'E0={endpoint:.1f}')
    plt.axvspan(*file_info["fit_range"], color='gray', alpha=0.2)
    plt.axhline(0, color='k', linestyle='--', alpha=0.5)
    plt.xlim(0, 900)  # Focus on relevant region
    plt.ylim(0, max(y)*1.1)
    plt.title(f'{thicknesses[i]}cm')
    plt.grid(True)

plt.tight_layout()
plt.show()

# Endpoint vs thickness plot
plt.figure(figsize=(8, 5))
plt.plot(thicknesses, endpoints, 'bo-')
plt.xlabel('Aluminum Thickness (cm)')
plt.ylabel('Endpoint Channel')
plt.title('Endpoint vs Thickness')
print(endpoints)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import curve_fit
plt.rcParams.update({'font.size': 12})
# Define the exponential function to fit the data
def exponential_func(x, a, b, c):
    return a * np.exp(-b * x) + c

# Load the data from the file
file_name = "day2/aufloesung/lin_fit.txt"
data = np.loadtxt(file_name, skiprows=1)

# Assuming the first column is DelE/Channel and the second is Channel
x = data[:, 0]
y = data[:, 1]

# Fit the data to the exponential function
popt, pcov = curve_fit(exponential_func, x, y, p0=[1, 1e-6, 1])

# Plot the original data
plt.scatter(x, y, label='Data')

# Plot the fitted curve
x_fit = np.linspace(min(x), max(x), 500)
y_fit = exponential_func(x_fit, *popt)
plt.plot(x_fit, y_fit, 'r-', label='Fitted exponential curve')

# Add labels and legend
plt.xlabel('Resolution')
plt.ylabel('Channel')
plt.title('Exponential Fit to Data')
plt.legend()
plt.savefig("day2/figs/aufloesung.png")
plt.show()

# Print the optimized parameters
print(f"Optimized parameters: a={popt[0]}, b={popt[1]}, c={popt[2]}")
```
